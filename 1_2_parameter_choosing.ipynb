```python
eps=[]
for i in range(fromvalue,tovalue,step):
  eps.append(i)
eps = np.array(eps)
eps.reshape(-1,1)
eps
```

```python

neigbors=np.vstack((eps, neigbors)).T
print(neigbors)
```

```python
neigbors_df=pd.DataFrame(neigbors,columns=['eps','total_count'])

#neigbors_df.to_csv('distances_xj.csv')
#from google.colab import files
#files.download('distances_xj.csv')
neigbors_df
```

```python
epsi=[]
epsilon=pd.DataFrame()
for i in range(fromvalue,tovalue,step):
  for j in range(781):
    epsi.append(i)
epsilon = pd.DataFrame(epsi, columns = ['eps'])
```

```python

count_df = pd.concat([count_df, epsilon], axis=1, join='inner')
```

```python
count_df
```

```python
total_df= pd.DataFrame()
#total_df=neigbors_xi_df.join(neigbors_df.set_index('b'), on='b', validate= None)
#total_df
total_df=count_df.merge(neigbors_df, left_on='eps', right_on='eps')
#total_df['neigbors_xj'] = total_df['neigbors_xj'].replace(0,1)
```

```python

total_df['total_count2'] = total_df['total_count'].replace(0,1)
total_df['xidividedbyxj'] = total_df.apply(lambda row: row['count'] / row['total_count2'], axis=1)
total_df
```

```python
total_df['xidividedbyxj2'] = total_df['xidividedbyxj'].replace(0,1)
total_df
```

```python

dfsai=pd.DataFrame()
import math

total_df['logdiv']=total_df.apply(lambda row: math.log2(row['xidividedbyxj2']), axis=1) 
total_df['forsummation']=total_df.apply(lambda row:row['xidividedbyxj']*row['logdiv']*(-1), axis=1) 

total_df
#dfsai = total_df.groupby('eps').sum('forsummation')

#dfsai=dfsai.drop(['count','total_count','xidividedbyxj','logdiv','id'], axis=1)
#dfsai
```

```python
total_df=total_df.drop(['xidividedbyxj2','logdiv'], axis=1)
#total_df.to_csv('total_distances.csv')
#from google.colab import files
#files.download('total_distances.csv')
```

```python
dfsai = total_df.groupby('eps').sum('forsummation')
dfsai = dfsai.reset_index(drop=False)
dfsai
```

```python
dfsailogeps=pd.DataFrame()
dfsailogeps['logeps']=dfsai.apply(lambda row: math.log2(row['eps']), axis=1) 
dfsailogeps
```

```python



dfsai=dfsai.drop(['count','total_count2','total_count','xidividedbyxj','tag'], axis=1)
dfsai
```
