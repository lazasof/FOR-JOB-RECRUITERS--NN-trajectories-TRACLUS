```python
# Input: (1) A cluster Ci of line segments, (2) M inLns, (3) A smoothing parameter γ
#Output: A representative trajectory RTRi for Ci
#Algorithm:
#01: Compute the average direction vector −→V ;
#02: Rotate the axes so that the X axis is parallel to −→V ;
#03: Let P be the set of the starting and ending points of the line segments in Ci;
#/* X`-value denotes the coordinate of the X`axis */
#04: Sort the points in the set P by their X`-values;
#05: for each (p ∈ P) do
#    /* count nump using a sweep line (or plane) */
#    06: Let nump be the number of the line segments that contain the X`-value of the point p;
#    
#    07: if (nump ≥ M inLns) then
#        08: diff := the difference in X`-values between p and its immediately previous point;
#        09: if (diff ≥ γ) then
#            10: Compute the average coordinate avg`p;
#            11: Undo the rotation and get the point avgp;
#            12: Append avgp to the end of RT Ri;
```

```python
import pandas as pd
import numpy as np
#header='infer', index_col=None
path="/content/drive/MyDrive/dfdist66037minlns4573.csv"
df1= pd.read_csv(path,header=0,index_col=0)
#Input: (1) A set of line segments D = {L1, · · · , Lnumln },####ALL CHARACTERISTIC LINE SEGMENTS\


tlon=df1['tlon']
tlat=df1['tlat']
df=df1.copy(deep=True)
df1['threelon']=df1['tlon']-tlon
df1['fourlon']=df1['Nlon']-tlon
df1['threelat']=df1['tlat']-tlat
df1['fourlat']=df1['Nlat']-tlat
dfcluster0=df1[df1['clusterId'] ==0]
dfcluster1=df1[df1['clusterId'] ==1]
dfcluster2=df1[df1['clusterId'] ==2]
print(len(df1))

dfcluster0 = dfcluster0.reset_index(drop=True)
dfcluster1 = dfcluster1.reset_index(drop=True)
dfcluster2 = dfcluster2.reset_index(drop=True)
```

```python
minlns=100
g=9000
import math
sum1=0
sum2=0
for i in range(len(dfcluster0)):
  sum1=sum1+dfcluster0.fourlon[i]
  sum2=sum2+dfcluster0.fourlat[i]
avglon=sum1/len(dfcluster0)
avglat=sum2/len(dfcluster0)
avgvector=np.array([[avglon],[avglat]])

avgvector

angle=math.atan2(avglat,avglon)

b=np.array([[avglon],[avglat]])
matrix= np.array([[math.cos(angle),(math.sin(angle))],[-(math.sin(angle)),math.cos(angle)]])
d=matrix@b
newmatrix= np.array([[math.cos(angle),-(math.sin(angle))],[(math.sin(angle)),math.cos(angle)]])
z=newmatrix@d
print(b)
print(d)
print(z)


c = np.array([])
d = np.array([])
for i in range(len(dfcluster0)):
  b=np.array([[dfcluster0.tlon[i]],[dfcluster0.tlat[i]]])
  z=np.array([[dfcluster0.Nlon[i]],[dfcluster0.Nlat[i]]])
  a=matrix@b
  y=matrix@z
  c = np.append(c,a) 
  d = np.append(d,y) 
print(c)  

c=np.reshape(c, (-1,2))   
d=np.reshape(d, (-1,2)) 
print(c)

dfcluster0[["newtlon", "newtlat"]] = c
dfcluster0[["newNlon", "newNlat"]] = d
dfcluster0

for i in range(len(dfcluster0)):
  if dfcluster0.newNlon[i]< dfcluster0.newtlon[i]:
    dfcluster0.loc[i, 'a'] = 11
    replace1=dfcluster0.newNlon[i]
    replace2=dfcluster0.newtlon[i]
    replace3=dfcluster0.newNlat[i]
    replace4=dfcluster0.newtlat[i]
    dfcluster0.loc[i, 'newNlon'] = replace2
    dfcluster0.loc[i, 'newtlon'] = replace1
    dfcluster0.loc[i, 'newNlat'] = replace4
    dfcluster0.loc[i, 'newtlat'] = replace3

df2 = dfcluster0.sort_values('newtlon')
df2=df2.reset_index(drop=True)
df2=df2[['newtlon','newtlat','newNlon','newNlat','line']]
firstrow=df2.iloc[0]
lastrow=df2.iloc[-1]
df2

import pandas as pd

print(df2)

rtr0=pd.DataFrame(columns=['newtlon','newtlat',' newNlon','newNlat','line'])
rtr0.loc[len(rtr0)]=firstrow

for i in range(1,len(df2)):
  sumlon=0
  sumlat=0
  sumNlon=0
  sumNlat=0
  nump=0
  #06: Let nump be the number of the line segments that contain the X`-value of the point p;
  point=df2.newtlon[i]
  for j in range(len(df2)):
    if point >= df2.newtlon[j] and point<= df2.newNlon[j]:
      nump=nump+1
      sumlon=sumlon+df2.newtlon[j]
      sumlat=sumlat+df2.newtlat[j]
      sumNlon=sumNlon+df2.newNlon[j]
      sumNlat=sumNlat+df2.newNlat[j]

  if nump >= minlns:
    checker=rtr0.loc[len(rtr0)-1]
    check=checker['newtlon']
    #diff := the difference in X`-values between p and its immediately previous point;
    diff=df2.newtlon[i]-check
    if abs(diff) >= g :
    #10: Compute the average coordinate avg`p;
      avglon=sumlon/nump
      avglat=sumlat/nump
      avgNlon=sumNlon/nump
      avgNlat=sumNlat/nump
      line=df2.line[i]
      rtr0.loc[len(rtr0)]=[avglon,avglat,avgNlon,avgNlat,3]
      rtr0=rtr0.reset_index(drop=True)
    #11: Undo the rotation and get the point avgp;
    #12: Append avgp to the end of RT Ri;




c = np.array([])
d = np.array([])
rtr0.loc[len(rtr0)]=lastrow
for i in range(len(rtr0)):
  b=np.array([[rtr0.newtlon[i]],[rtr0.newtlat[i]]])
  a=newmatrix@b
  c = np.append(c,a) 
print(c)  
c=np.reshape(c, (-1,2))   
print(c)
rtr0[["newlo", "newla"]] = c




rtr0 = rtr0.sort_values(by = 'newlo')
rtr0

import matplotlib.pyplot as plt

fig,ax=plt.subplots()
ax.plot(dfcluster0['tlon'],dfcluster0['tlat'], '-o', label='characteristic Line segments cluster 0')
ax.plot(rtr0['newlo'],rtr0['newla'], '-x', label='representative')
ax.legend()
plt.show()
```
